```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal

# Read txt file
with open('data/random.txt', 'r') as txtfile:
    txtlines = txtfile.read().splitlines()
disp_operation = np.array(txtlines, dtype=int)

# Read CSV file
header_rows1 = 0
df1 = pd.read_csv('data/fixed/bin_data_fixed_0.csv', skiprows=header_rows1, low_memory=False)
disp_result = df1.iloc[0][1:].to_numpy()
```

```python
plt.scatter(disp_operation, disp_result, marker='o', s=1)

# 添加标题和标签
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 显示图形
plt.show()
```
